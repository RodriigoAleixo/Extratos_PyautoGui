```python
# Código Completo

import pyautogui
import time
import openpyxl

#Abrir navegador
pyautogui.press('win')
time.sleep(1)
pyautogui.write('Chrome')
time.sleep(1)
pyautogui.press('Enter')
time.sleep(2)
#pyautogui.hotkey('win', 'up')
time.sleep(1)

#Acessar NewCon
pyautogui.write('https://newcon.bancorbras.com.br/intranet/')
time.sleep(1)
pyautogui.press('Enter')
time.sleep(2)
pyautogui.press('tab', presses=2)
time.sleep(1)
pyautogui.press('Enter')
time.sleep(3)

#Unidade de negócio
x, y, largura, altura = pyautogui.locateOnScreen('unidade.png', grayscale= True, confidence= 0.7)
time.sleep(1)
pyautogui.click(x + largura/2, y + altura/2)
time.sleep(3)

#Importanto a planilha.
workbook = openpyxl.load_workbook('extrato.xlsx')
#Indicando qual aba eu quero utilizar.
pagina_clientes = workbook['Plan1']
#Informando qual a primeira linha de leitura
for linha in pagina_clientes.iter_rows(min_row=2):
    grupo = linha [0].value
    cota = linha[1].value
    nome = linha[2].value

    #Atendimento
    x, y, largura, altura = pyautogui.locateOnScreen('atendimento.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(1)

    #Grupo:
    time.sleep(3)
    pyautogui.write(str(grupo))
    time.sleep(1)
    pyautogui.press('tab')
    time.sleep(2)

    #Cota:
    pyautogui.write(str(cota))
    time.sleep(1)
    pyautogui.press('tab', presses=2)
    time.sleep(1)
    pyautogui.press('Enter')
    time.sleep(6)
    pyautogui.press('tab', presses=20)
    time.sleep(1)
    pyautogui.press('Enter')
    time.sleep(1)
    pyautogui.scroll(-500)
    time.sleep(2)   

    #Mensagens
    x, y, largura, altura = pyautogui.locateOnScreen('mensagens.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(3) 
    pyautogui.scroll(-500)
    time.sleep(2) 

    #Imprimir
    x, y, largura, altura = pyautogui.locateOnScreen('imprimir.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(2)

    #Extrato Atualizado
    x, y, largura, altura = pyautogui.locateOnScreen('extrato_atualizado.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(4    )
    pyautogui.press('esc')
    time.sleep(3)
    

    #Imprimir
    pyautogui.hotkey('win', 'up')
    time.sleep(5)
    pyautogui.press('tab', presses=6)
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(4)

    #Renomear Arquivo
    pyautogui.write(str(grupo))
    time.sleep(1)
    pyautogui.write('-')
    time.sleep(1)
    pyautogui.write(str(cota))
    time.sleep(1)
    pyautogui.press('enter')
    time.sleep(4)

    #Fechar
    pyautogui.hotkey('alt', 'f4')
    time.sleep(2)
    x, y, largura, altura = pyautogui.locateOnScreen('fechar.png', grayscale= True, confidence= 0.7)
    time.sleep(1)
    pyautogui.click(x + largura/2, y + altura/2)
    time.sleep(1)
```
